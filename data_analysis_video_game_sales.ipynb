```python
df["jp_sales"] = df["JP_Sales"].apply(lambda x : x*1000000)
df.drop("JP_Sales", inplace = True, axis = 1)

df["na_sales"] = df["NA_Sales"].apply(lambda x : x*1000000)
df.drop("NA_Sales", inplace = True, axis = 1)

df["eu_sales"] = df["EU_Sales"].apply(lambda x : x*1000000)
df.drop("EU_Sales", inplace = True, axis = 1)

df["global_sales"] = df["Global_Sales"].apply(lambda x : x*1000000)
df.drop("Global_Sales", inplace = True, axis = 1)

df["other_sales"] = df["Other_Sales"].apply(lambda x : x*1000000)
df.drop("Other_Sales", inplace = True, axis = 1)

df.head()
```

```python
plt.figure(figsize = (20, 10))
sns.barplot(x = "na_sales", y = "Genre", data = df)
plt.title("Sales of Genre in North America")
```

```python
plt.figure(figsize = (20, 10))
sns.barplot(x = "jp_sales", y = "Genre", data = df)
plt.title("Sales of Genre in Japan")
```

```python
plt.figure(figsize = (20, 10))
sns.barplot(x = "global_sales", y = "Genre", data = df)
plt.title("Sales of Genre in Globally")
```

```python
most_popular = df.query('Rank<51', inplace = False).sort_values('Rank', ascending = False)
most_popular
```

```python
na = df["na_sales"].sum()
jp = df["jp_sales"].sum()
eu = df["eu_sales"].sum()
os = df["other_sales"].sum()
df["global_sales"].sum()
```

```python
# how much sales in different reigon contribute to global sales
plt.figure(figsize = (20, 10))
y = np.array([na, jp, eu, os])
mylabels = ["Sales in North America", "Sales in Japan", "Sales in European Union", "Sales in Other Reigons"]
plt.pie(y, labels = mylabels)
plt.show() 
```

```python
print("Sales of top 20 games of all time in different reigons")
for i in range(0, 21):
    na = df["na_sales"][i]
    jp = df["jp_sales"][i]
    eu = df["eu_sales"][i]
    os = df["other_sales"][i]
    
  
    plt.figure(figsize = (20, 10))
    y = np.array([na, jp, eu, os])
    mylabels = ["Sales in North America", "Sales in Japan", "Sales in European Union", "Sales in Other Reigons"]
    plt.title(f"Sales of {df['Name'][i]}")
    plt.pie(y, labels = mylabels)
    plt.show() 
```
